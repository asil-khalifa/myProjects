# Histograms

### np.random.normal(loc=0.0, scale=1.0, size=None)
this draws random samples from a gaussian distribution

loc: mean of this distribution<br>
scale: standard deviation<br>
size: no of values to return

### plt.hist() is used for histogram

bins =<br>
    a. integer: then there will that many equally sized bars<br>
    b. list [a, b, c, d]: then there will bins like [a, b) [b, c) [c, d] (Only last will be closed on both sides)

cumulative = True gives you a cumulative graph

```python
x = np.random.normal(20, 1.5, 1000)
plt.hist(x, bins=25, cumulative=True) #[x.min(), 18, 21, x.max()]
```

# Pie Chart

### plt.pie()

labels = []<br>
explode = [], this allows you to add a number (fraction of radius) of explosion to each label, causing it to come out<br>
autopct allows you to add percentage<br>
pctdistance sets the distance of the percentage label from the center<br>
startangle is anti-clockwise, from the right (like sin, cos circle)



```python
langs = ['python', 'c++', 'c', 'javascript', 'java']
votes = [20, 50, 24, 14, 6]
explodes = [0.2, 0, 0, 0, 0]
plt.pie(votes, labels=langs, explode=explodes, autopct='%.2f%%', pctdistance=.65, startangle=0)
```

# Plot Customization

<ul>

# Multiple Figures

The below is one way of doing it
<br>Here, there are two separate images (in .py files, two separate windows)

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)
plt.figure(1)
plt.scatter(x1, y1)
plt.figure(2)
plt.plot(x2, y2)

```

You do this by writing<br>
fig, axs = plt.subplots(nrows, ncols)<br>
You can access each subplot using axs[0][0] or axs[0, 0], and so on<br>
you have to use: <b>axs[0, 0].set_title()</b><br>
<b>fig.suptitle()</b> is used to set the overall title<br>
<b>plt.tight_layout()</b> prevents overlap of labels between various subplots

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sine curve')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cosine curve')

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title('random function')

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title('log function')

fig.suptitle('Four Plots')
```

# Exporting Plots

<b>plt.savefig()</b> is used to save a figure
<ul>
<li><b>dpi</b> parameter is used to set quality of the produced image</li>
<li><b>transparent</b> parameter set to <b>True</b> is used to remove the white background</li>
</ul>

```python
x = np.arange(100)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('sine curve')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('cosine curve')

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title('random function')

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title('log function')

fig.suptitle('Four Plots')
plt.tight_layout()

plt.savefig('four_plots.png', dpi = 1000, transparent = False)
```

# 3d Plotting

<b>ax = plt.axes(projection = '3d')</b><br>
This allows you to add axes to the plot and returns this axes

```python
ax = plt.axes(projection = '3d')

x, y, z = np.random.random(100), np.random.random(100), np.random.random(100)

ax.scatter(x, y, z)
ax.set_title('3D Plot')
```

```python
ax = plt.axes(projection = '3d')

x = np.arange(0, 50, .1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title('3D Plot')
```

### To create a surface Plot:

See geeks for geeks to understand np.meshgrid()

<b>cmap</b> parameter adds depth dependent color to the surface plot<br>
<b>ax.view_init(azim, elev)</b> allows you to set the initial view

```python

ax = plt.axes(projection = '3d')
x = np.arange(-5, 5, .1)
y = np.arange(-5, 5, .1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap = 'viridis')
```

# Animating Plots

<b>plt.pause()</b> allows you to animate the graph

```python
%matplotlib qt

heads_tails = [0, 0]
for _ in range(100_000):
    heads_tails[random.randint(0, 1)] += 1
    plt.bar(['heads', 'tails'], heads_tails, color = ['r', 'b'])
    plt.pause(.001)
    
%matplotlib inline
```
